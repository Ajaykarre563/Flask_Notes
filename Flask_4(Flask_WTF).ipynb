# **What is Flask-WTF?**

**Flask-WTF(Web Template Forms) is an extension for Flask that helps you create and manage web forms easily. It is built on top of WTForms and adds extra features like form validation, CSRF protection, and handling user input securely.**

# **Why use Flask-WTF?**

# **Simplifies Form Handling in Flask**

Without Flask-WTF:
You need to manually handle form data using request.form, check if fields are filled, and validate them.

With Flask-WTF:
It automatically handles form inputs, validation, and security, making the code cleaner.

# **Provides Built-in Validation**

Validation ensures that the user enters the correct data (e.g., required fields, proper email format).

# **Prevents CSRF Attacks with a Security Token**

CSRF (Cross-Site Request Forgery) is a security attack where a hacker tricks a user into submitting a form without their knowledge.

Flask-WTF automatically adds a secret security token to forms to prevent this.

# **Summary**

**TextAreaField:** Multi-line text input (e.g., message).

**IntegerField:** Integer input (e.g., age).

**SubmitField:** Submit button.

**RadioField:** buttons (e.g., gender selection).

**SelectField:** Dropdown menu (e.g., language selection).

**validators:** Used to validate user input (e.g., ensuring a field is not empty or is an email).

**ValidationError:** Custom error raised when a field fails validation.

# **What is session in Flask?**

**In Flask, session is a way to store data across multiple requests for a user. It allows you to save user-specific data (like login details) so that the user doesn't have to re-enter them on every page.**

**Simple Example:**

Imagine you log in to a website, and after logging in, the site remembers your username on every page. Flask achieves this using session.

# **Advantages of Using Session in Flask**

**Stores User Data Temporarily**

Keeps track of user-specific information (e.g., login status, preferences) across multiple pages.

**Secure Data Storage**

**Lightweight & Fast**

Uses a dictionary-like structure (session['key'] = value), making it easy to store and retrieve data quickly.

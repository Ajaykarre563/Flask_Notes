# **what is mean by Flask ?**

**Flask is a web framework for Python that helps you build web applications. Think of it as a toolkit that makes it easier to create websites or web services by providing a set of simple tools handle user requests, and send back responses, like showing a page or handling a form submission.**

# **Flask is used to:**

Build websites and apps.

Show pages, process forms, and connect to databases.

Create APIs for different software to communicate.

## **drawbacks of Flask**

**Limited Features:**Flask has fewer built-in tools, so you may need to add extra libraries for complex tasks.

**Not for Big Projects:** It might not handle very large apps as well as other frameworks like Django.

**More Setup:** You need to set up things like login and databases manually, which can take time.

**Handles Fewer Users:** Flask works best with fewer users at once; it needs extra work to handle lots of traffic.

**No Built-in Admin Panel:** Unlike other frameworks, Flask doesn't have a ready-made admin dashboard.

# **Essential built-in functions in Flask, along with their explanations and examples.**

**1.Flask()**
Creates a Flask web application instance.



```
from flask import Flask
app = Flask(__name__)

```



# **What is Routing?**

Routing is the mechanism by which Flask determines how to respond to a client request based on the URL. Each route is associated with a specific function that is executed when the route is accessed.

**2.route():Defines the URL for a function.**



```
@app.route("/")
def home():
    return "Hello, Flask!"

```



# **Dynamic Routing**
Flask also supports dynamic routing, allowing you to capture values from the URL. For example:



```
@app.route('/user/<username>')
def show_user_profile(username):
    return f'User: {username}
```



3.**run()**:Starts the Flask web server.



```
if __name__ == "__main__":
    app.run(debug=True)

```



4.**render_template():**Renders an HTML file from the templates folder.



```
from flask import render_template
@app.route("/home")
def home():
    return render_template("home.html")

```



**5.url_for():**Generates a URL for a given function.



```
from flask import url_for
@app.route("/home")
def home():
    return "Home Page"

@app.route("/redirect_home")
def go_home():
    return redirect(url_for("home"))


```



# **Note:**

**In Flask, you can pass a Python file (or data from it) to an HTML template using context (conventionally called context or data) in render_template.**



```
from flask import Flask, render_template

app = Flask(__name__)  # Create a Flask instance

# Sample data (this can come from a Python file)
import data  # Import the Python file

@app.route("/")
def home():
    return render_template("home.html", content=data.message)  # Pass data to HTML

if __name__ == "__main__":
    app.run(debug=True)

```

# **what is mean by HTTP?**

**In Flask, HTTP (HyperText Transfer Protocol) is the communication system used by web applications to send and receive data between a client (browser) and a server (Flask app).**

# **Simple Explanation:**

When you visit a website, your browser sends an HTTP request to the server.
The Flask app receives the request, processes it, and sends an HTTP response back to the browser.

**Example in Flask:**

GET Request (Fetching a page)
When you visit http://localhost:5000/home, Flask handles the GET request and returns home.html.

POST Request (Submitting a form)
When you submit a login form, the browser sends a POST request with your data (e.g., username). Flask processes it and responds accordingly.
Key HTTP Methods in Flask:

 GET  Retrieve data (e.g., load a webpage)

 POST  Send data (e.g., submit a form)

 PUT  Update data

 DELETE  Remove data

# **Handling HTTP Methods**

**Flask allows you to handle different HTTP methods such as GET, POST, PUT, and DELETE. By default, routes respond to GET requests. To specify other methods, you can use the methods parameter in the route decorator:**



```
@app.route('/submit', methods=['POST'])
def submit_form():
    return 'Form submitted!'
```



**In this example, the submit_form function will only be called when a POST request is made to the /submit route.**

# **Request and Response Objects**

**Flask provides request and response objects to handle incoming data and send responses back to the client.**

# **The Request Object**

**The request object contains all the information about the incoming request, including form data, query parameters, and headers. Heres an example of how to access form data:**



```
from flask import request

@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']
    password = request.form['password']
    return f'Username: {username}, Password: {password}'
```



# **The Response Object**

**You can return various types of responses from your Flask routes, including strings, HTML, JSON, and more. For example, to return JSON data, you can use the jsonify function:**



```
from flask import jsonify

@app.route('/data')
def data():
    return jsonify({'key': 'value'})
```
